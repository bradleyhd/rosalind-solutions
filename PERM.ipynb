```python
import math
import itertools

n = 7;
print(len(list(itertools.permutations(range(1,n+1)))))
for p in itertools.permutations(range(1,n+1)):
    for x in range(0, len(p)):
        if(x==len(p)-1):
            print(p[x], end="")
        else:
            print(str(p[x]) + " ", end="")
    print("")
```
